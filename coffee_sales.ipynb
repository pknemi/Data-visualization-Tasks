```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv("coffee_sales.csv")
df
```

```python
df.columns
```

**1.ANALYSIS OF DIFFERENT TYPES OF PRODUCT CATEGORIES**

```python
pc=df["product_category"].value_counts()
pc
```

```python
plt.figure(figsize=(10,8))
wedgeprops={"edgecolor":"black","linewidth":3}

plt.pie(pc,labels=pc.index,wedgeprops=wedgeprops,autopct="%1.1f%%")
plt.title("POPULAR PRODUCT CATOGORIES")
plt.legend(bbox_to_anchor=(1,1))
```

INSIGHT-Coffee is the most popular product category followed by Tea

**2.ANALYSIS OF TYPES OF PRODUCTS**

```python
p_type=df["product_type"].value_counts()
p_type
```

```python
plt.figure(figsize=(10,10))
plt.bar(p_type.index,p_type.values,color="brown")
plt.title("ANALYSIS OF DIFFERENT TYPES OF PRODUCTS")
plt.xlabel("PRODUCT TYPE")
plt.ylabel("COUNT")
plt.xticks(rotation=90)
```

INSIGHT-Brewed chai tea is the most popular product followed by Gourmet brewed coffee.

**3.ANALYSIS OF  DAILY SALES TRENDS**

```python
daily_sales = df.groupby(df['transaction_date']).size()
d_sales=daily_sales.sort_values(ascending=False).head()
d_sales
```

```python
plt.style.use("dark_background")
plt.plot(d_sales.index,d_sales.values,"r*--")
plt.title("DAILY SALES TRENDS")
plt.xlabel("DATE")
plt.ylabel("TRANSACTION COUNT")
plt.xticks(rotation=90)
```

INSIGHT-The most number of transaction held on 2023-06-19,and the least was on 2023-06-27

```python

monthly_sales = df.groupby(df['Month Name']).size()
monthly_sales.sort_values()#reindex(['January', 'February', 'March', 'April', 'May', 'June'])
```

**4.MONTHLY TRANSACTION ANALYSIS**

```python
plt.title("MONTH WISE TRANSACTIONS")
plt.ylabel("TRANSACTION COUNT")
sns.lineplot(x=monthly_sales.index,y=monthly_sales.values,marker="*",markerfacecolor="red",markersize=10)
```

INSIGHT-In the first 6 months of an year,it can be seen that most number of transactions was held in june followed by may

```python
store_sales = df['store_location'].value_counts()
store_sales
```

**5.ANALYSIS OF LOCATION WISE SALES TRENDS**

```python
plt.bar(store_sales.index,store_sales.values,color="white")
plt.title("DISTRIBUTION SALES ACROSS LOCATION")
plt.xlabel("LOCATION")
plt.ylabel("TRANSACTIONS")
```

INSIGHT-Higher sales occured at Hell's kitchen followed by Astoria.

**6.TOP 5 SELLING PRODUCTS**

```python
p_type=df["product_type"].value_counts()
top_p_type=p_type.head()
top_p_type
```

```python
plt.pie(top_p_type,labels=top_p_type.index,autopct="%1.1f%%")
plt.title("TOP 5 SELLING PRODUCTS")
plt.legend(bbox_to_anchor=(1,1))
```

INSIGHT-Brewed Chai tea leads, closely followed by Gourmet brewed coffee.

**7.DISTRIBUTION OF TRANSACTION QUANTITY**

```python
transaction_qty_dist = df['transaction_qty']
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(transaction_qty_dist, bins=8, color='LIGHTGREEN')
plt.title('Distribution of Transaction Quantities')
plt.xlabel('Transaction Quantity')
plt.ylabel('Frequency')
plt.show()
```

INSIGHT-The histogram shows that most of  the transactions involve 1 or 2 items, with a significant decline in frequency for higher quantities.

**8.Sales Trends by Day of the Week**

```python
sales_by_day = df.groupby(df["Day Name"]).size()
sbd=sales_by_day.sort_values()
sbd
```

```python
plt.style.use("dark_background")
colors=["violet","indigo","blue","green","yellow","orange","red"]
plt.bar(sbd.index,sbd.values,color=colors)
plt.title("SALES TRENDS BY DAY")
plt.xlabel("DAY")
plt.ylabel("COUNT")
plt.xticks(rotation=90)
```

INSIGHT-The bar chart shows a fairly consistent number of transactions throughout the week,with a slight change on thursday and fridays.

**9.FINDING CORRELATION AMONG TRANSACTION QUANTITY,TOTAL BILL, UNIT PRICE**

**10.LOCATION WISE DISTRIBUTION OF PRODUCT CATEGORY**

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='product_category',hue="store_location",data=df,palette="pastel")
plt.title('Location-wise Product Category Count Distribution')
plt.xlabel('Product Category')
plt.xticks(rotation=90)
plt.show()
```

**11.TOTAL SALES IN A MONTH**

```python

Total_sales_month = df.groupby('Month Name')['Total_Bill'].sum().reindex(['January', 'February', 'March', 'April', 'May', 'June'])
Total_sales_month
```

```python
plt.figure(figsize=(8, 6))
plt.plot(Total_sales_month, marker='o', color='lightskyblue')
plt.title('Total Sales Trend Over the Months')
plt.xlabel('Month')
plt.ylabel('Total Sales')
```

INSIGHT-The line graph displays a clear upward trend in total sales from January to June.
