```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("car_prices.csv")
```

```python
df.shape
```

```python
df
```

```python
df.shape
```

```python
high_selling_price=df.sort_values(by="sellingprice",ascending=False).head(10)
```

```python
high_mmr=df.groupby("make")["mmr"].sum().sort_values(ascending=False).head(10)
```

```python
plt.style.use("ggplot")
plt.bar(high_selling_price["make"],high_selling_price["sellingprice"],color="pink")
plt.title("Cars with high selling price")
plt.xlabel("cars")
plt.ylabel("selling price")
```

**INSIGHT-Ferrari having the highest selling price followed by Rolls-Royce**

```python
condition_vehicles=df.sort_values(by="condition",ascending=False).head(10)
condition_vehicles_name=condition_vehicles["make"].head(10)
Top_condition_vehicles=condition_vehicles["condition"]
plt.bar(condition_vehicles_name,Top_condition_vehicles,color="cyan")
plt.xticks(rotation=90)
plt.xlabel("Car Brand")
plt.ylabel("Condition Rating")
plt.title("TOP 10 BRANDS WITH BEST CONDITION")
```

INSIGHT-Most of the brands have good condition rating

```python
plt.style.use("classic")
high_mmr.plot(kind="bar",x=high_mmr.index,y=high_mmr.values,color="red")
plt.title("Marketing value of different brands")
plt.xlabel("cars")
plt.ylabel("marketing price")
```

**INSIGHT-Ford brand has the highest marketing value followed by cheverolet**

```python
low_mmr=df.groupby("make")["mmr"].sum().sort_values(ascending=False).tail(10)
low_mmr
```

```python
plt.style.use("dark_background")
low_mmr.plot(kind="line",x=low_mmr.index,y=low_mmr.values,color="brown",lw=5)
plt.xticks(rotation=90)
plt.title("brands having lowest marketing price")
plt.ylabel("marketing price")
```

**INSIGHT-Ford truck is the one with lowest marketing price succeeded by dodge tk**

```python
gear_group=df["transmission"].value_counts().head(10)
gear_group
```

```python
wedgeprops={"edgecolor":"black","linewidth":5}
plt.pie(gear_group,labels=gear_group.index,wedgeprops=wedgeprops,autopct="%f1.1f%%")
plt.title("Transmission distribution")
```

**INSIGHT-95% of the users use Automatic gear**

```python

plt.style.use("dark_background")
plt.scatter(df['odometer'],df['sellingprice'],color='white')
plt.title('RELATION BETWEEN KILOMETERS COMPLETED AND THEIR SELLING PRICES')
plt.xlabel('KILOMETERS')
plt.ylabel('SELLING PRICE')
```

INSIGHT-Selling price is inversely proportional to kilometers

```python
car_group=df.groupby("make")
kia_cars=car_group.get_group("Kia").head(10)
kia_cars_interior_color=kia_cars["interior"].value_counts()
plt.pie(kia_cars_interior_color,labels=kia_cars_interior_color.index,autopct="%1.1f%%")
plt.title("Kia interior colors")
plt.legend(bbox_to_anchor=(1.5,1))
```

```python
plt.style.use("dark_background")
year_group=df.groupby("year")
cars_2014=year_group.get_group(2014).head(10)
car_colors_2014=cars_2014["color"]
car_names_2014=cars_2014["make"]
plt.scatter(car_names_2014,car_colors_2014)
plt.title("car manufacturing by color in 2014")
plt.xlabel("Car Brand")
plt.ylabel("Car colors")
```

INSIGHT-In 2014,blue was the most commonly manufactured color for cars

```python
year_group=df.groupby("year")
cars_2013=year_group.get_group(2013).sort_values(by="sellingprice",ascending=False)
Top_sellingprice_cars_2013=cars_2013["sellingprice"].head(10)
Top_sellingprice_car_names_2013=cars_2013["make"].head(10)
plt.bar(Top_sellingprice_car_names_2013,Top_sellingprice_cars_2013,color="magenta")
plt.title("2013:Car Manufacturing  and Price Analysis")
plt.xlabel("Car Brand")
plt.ylabel("Selling Price")
plt.xticks(rotation=90)
```

INSIGHT-In 2013,Bentley had the highest selling price among car manufacturers

```python
car_group=df.groupby("make")
bmw_cars=car_group.get_group("BMW").head(10)
bmw_car_colors=bmw_cars["color"].value_counts()
plt.pie(bmw_car_colors,labels=bmw_car_colors.index,autopct="%1.1f%%")
plt.title("BMW car colors")
plt.legend(bbox_to_anchor=(1.5,1))
```
