```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path="/content/makeup_product_data.csv"
```

```python
df=pd.read_csv(path)
df
```

```python
df.info()
```

```python
#1
sns.barplot(x=df["Product type"],y=df["Availability"],data=df,color="lightblue")
plt.title("Products and their Availability")
```

Insights:-
As per the graph,cosmetics and skincare products have greatest availability

```python
#2
sns.barplot(x=df["Product type"],y=df["Number of products sold"],data=df,color="brown")
plt.title("Products Vs No Of Product Sold")
```

Insights:-
As per the graph,Skincare product is highest sold product

```python
#3
sns.displot(df["Stock levels"],bins=5)
plt.title("Stock Distribution")
```

Insights:-
As per the graph,stock levels of products highest in between 0-20

```python
#4
sns.lineplot(x=df["Product type"],y=df["Revenue generated"],data=df,errorbar=None,marker="o",markersize=15,markerfacecolor="lightgreen")
plt.title("Products Vs Revenue")
```

Insights:-
As per the graph,cosmetics products are the highest revenue generated products

```python
#5
company_location=df['Location'].value_counts()
company_location
```

```python
plt.pie(company_location,labels=company_location.index,autopct="%1.1f%%")
plt.legend(title="company_location.index",bbox_to_anchor=(1.5,1))
```

Insights:-
Most of the supplier companies located in kolkata with 25.0%

```python
#6
inspection=df["Inspection results"].value_counts()
inspection
```

```python
plt.pie(inspection,labels=inspection.index,autopct="%1.1f%%",colors=["lightgreen","indigo","purple"])
plt.legend(title="inspection.index",bbox_to_anchor=(1.5,1))
```

Insights:-
As per the graph,Only 23.0% of the product passed the inspection test

```python
#7
Transportation=df["Transportation modes"].value_counts()
Transportation
```

```python
plt.pie(Transportation,labels=Transportation.index,autopct="%1.1f%%",colors=["red","yellow","green","violet"])
plt.legend(title="Transportation.index",bbox_to_anchor=(1.5,1))
plt.title("Transportation")
```

Insights:-
Most of the products transportation are Road and Rail
